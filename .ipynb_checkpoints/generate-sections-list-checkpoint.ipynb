```python
from bs4 import BeautifulSoup
```

```python
html = open('cl-homepage.html')
parsed = BeautifulSoup(html, "lxml")
```

```python
sections = parsed.findAll('div',{'class':'cats'})
for section in sections:
    print "\n"
    for x in section.findAll('a'):
        try:
            print '"'+x.text+'":"'+x['data-cat']+'",'
        except:
            continue
```
